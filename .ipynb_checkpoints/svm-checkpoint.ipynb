```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris.head()
```

```python
iris['species'].unique()
```

**Observe clear groupings in pairplot**

```python
sns.pairplot(iris,hue='species',diag_kind='hist')
```

```python
for species in iris['species'].unique():
    spe_ser = iris[iris['species']==species]
    sns.kdeplot(spe_ser['sepal_width'],spe_ser['sepal_length'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.drop('species',axis=1),iris['species'],test_size=0.3,random_state=101)
```

**Testing SVM Classifier**

```python
from sklearn.svm import SVC
```

```python
svc = SVC(gamma='auto')
```

```python
svc.fit(X_train,y_train)
```

```python
predictions = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

**Gridsearch Parameter Testing**

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 
```

```python
grid = GridSearchCV(SVC(),param_grid,verbose=2)
```

```python
grid.fit(X_train,y_train)
```

```python
best = grid.best_estimator_
preds = best.predict(X_test)
```

```python
print(confusion_matrix(y_test,preds))
```

```python
print(classification_report(y_test,preds))
```

**Random Forest Performance Comparison**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_preds = rfc.predict(X_test)
```

```python
print("Random Forest:\n", classification_report(y_test, rfc_preds))
print("SVM:\n", classification_report(y_test, preds))
```

**Conclusion: Both work very well, trends in training data observed in above graphs are apparent and easily learned by SVM/RFC classifiers**
